```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('datosQS-5000.csv')

tamanio = data['Tamanio']

mejor_caso = data['Mejor Caso']
peor_caso = data['Peor Caso']
caso_promedio = data['Caso Promedio']

plt.figure(figsize=(10, 6))

plt.plot(tamanio, mejor_caso, label='Mejor Caso', marker='o')
plt.plot(tamanio, peor_caso, label='Peor Caso', marker='s')
plt.plot(tamanio, caso_promedio, label='Caso Promedio', marker='o')

plt.title('Complejidad Temporal para 5000 elementos (QuickSort)')
plt.xlabel('Tamao del Arreglo')
plt.ylabel('Tiempo')

# Mostrar leyenda
plt.legend()

plt.grid(True)

plt.show()
```
